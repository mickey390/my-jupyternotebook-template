# initial setting

## setting my access tokens

- setting access tokens for Microsoft Azure Machine Learning Studio Notebooks
- check your workspece id & token on Microsoft Azure Machine Learning Studio  Settings
- see https://github.com/Azure/Azure-MachineLearning-ClientLibrary-Python#specify-workspace-via-config

### create settings.ini on local pc 

$ vi settings.ini

```
[workspace]
id=xxxxxxxxxxxx
authorization_token=xxxxxxxxx
api_endpoint=https://asiasoutheast.studioapi.azureml.net
management_endpoint=https://management.azureml.net
```

### upload settings.ini to notebook instance
